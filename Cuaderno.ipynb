### Introducción

#### Conceptos

##### La minería de datos (Data Mining) es el proceso de descubrir patrones, tendencias y conocimientos significativos a partir de grandes conjuntos de datos. Combina métodos de estadísticas, inteligencia artificial y aprendizaje automático para analizar la información desde diferentes perspectivas y resumirla en información útil que puede ser usada para aumentar ingresos, reducir costos o ambas.


####Ciencia de Datos

##### La Ciencia de Datos y la Minería de Datos son campos estrechamente relacionados, pero con enfoques ligeramente diferentes. Mientras que la minería de datos se centra en descubrir patrones y conocimientos dentro de grandes conjuntos de datos, la ciencia de datos es un campo más amplio que abarca todo el ciclo de vida de los datos, desde la adquisición y limpieza hasta el análisis, modelado, interpretación y comunicación de resultados. En este contexto, la minería de datos puede considerarse una técnica fundamental utilizada dentro del marco más amplio de la ciencia de datos para extraer valor de los datos.

#### Estadistica Basica

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado


```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

Mediana

```python
mediana = np.median(data)
print(mediana)
```

Varianza y Desviación Estándar

```python
varianza = np.var(data)
desviacion = np.std(data)
```

```python
print(varianza)
print(desviacion)
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000)
print(np.mean(muestra), np.std(muestra))
```

Correlación

```python
corr = np.corrcoef([1,2,3], [1,5,7])
print(corr)
```
